```python
conditions = [
    (df['Sex']=='female') & (df['Pclass']==1),
    (df['Sex']=='female') & (df['Pclass']!=1),
    (df['Sex']=='male') & (df['Pclass']==1),
    (df['Sex']=='male') & (df['Pclass']!=1)
]
results = [
    'F-High',
    'F-Low',
    'M-High',
    'M-Low'
]
%timeit df['Gender_Status'] = np.select(conditions,results,'Other')
```

```python
def family_size(val):
    if val>0:
        return False
    return True
```

```python
family_size(0)
```

```python
df['Total_Family_Size'].apply(family_size)
```

```python
def gender_status(row):
    if row['Sex']=='female' and row['Pclass']==1:
        return 'F-High'
    elif row['Sex']=='female' and row['Pclass']>1:
        return 'F-Low'
    elif row['Sex']=='male' and row['Pclass']==1:
        return 'M-High'
    elif row['Sex']=='male' and row['Pclass']>1:
        return 'M-Low'
    else:
        return 'Other'
```

```python
%timeit df.apply(gender_status,axis=1)
```
